```python
import numpy as np
```

```python
L1 = 5.71
L2 = 2.915
L3 = 1.486
T1_MIN = -46
T1_MAX = 66
T2_MIN = -174.5
T2_MAX = -51
T3_MIN = -141.5
T3_MAX = 31.5
CHANDOU_ANGLE = 52.47
```

```python
def EndPoint(theta1, theta2, theta3):
    x = L1*np.cos(np.pi/180*theta1) + L2*np.cos(np.pi/180*theta1 + np.pi/180*theta2) + L3*np.cos(np.pi/180*theta1 + np.pi/180*theta2 + np.pi/180*theta3)
    y = L1*np.sin(np.pi/180*theta1) + L2*np.sin(np.pi/180*theta1 + np.pi/180*theta2) + L3*np.sin(np.pi/180*theta1 + np.pi/180*theta2 + np.pi/180*theta3)
    return (x, y)
```

```python
print(EndPoint(0, -51, 31.5))
```

```python
Mat = np.zeros((100, 100), dtype = np.int)
```

```python
theta1 = T1_MIN
theta2 = T2_MIN
theta3 = T3_MIN
while theta1 <= T1_MAX:
    theta2 = T2_MIN
    while theta2 <= T2_MAX:
        theta3 = T3_MIN
        while theta3 <= T3_MAX:
            x, y = EndPoint(theta1, theta2, theta3)
            x_index = int(x/0.2) + 50
            y_index = int(y/0.2) + 50
            Mat[y_index][x_index] = 1
            theta3 += 0.5
        theta2 += 0.5
    theta1 += 0.5
    print("Now updating for theta1 = {}".format(theta1))
```
